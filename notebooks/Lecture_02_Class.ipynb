## 2. Boolean Expressions and Conditionals

- Comparisons: ==, !=, <, >, <=, >=
- Logical: and, or, not
- if / elif / else; membership with `in`

```python
a, b = 7, 3
print('a == b:', a == b)
print('a != b:', a != b)
print('a > b:', a > b)
print('a <= b:', a <= b)
print('(a > 5) and (b < 5):', (a > 5) and (b < 5))
print('(a > 5) & (b < 5):', (a > 5) & (b < 5))
print('(a < 5) or (b < 2):', (a < 5) or (b < 2))
print('not (a == 7):', not (a == 7))
```

 # Conditional Statements

```python
# if statement — basic condition
#Runs a block of code only if the condition is true.
x = 10

if x > 5:
    print("x is greater than 5")
    print("This line is also inside the if block")
```

```python
# if–else statement — two-way decision
x = 3

if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```

```python
# if–elif–else chain — multiple conditions
score = 75

if score >= 90:
    print("The grade is A")
elif score >= 80:
    print("The grade is B")
elif score >= 70:
    print("The grade is C")
else:
    print("The grade is F")
```

## 3. Loops: for, while, range()

- for over lists/strings/dicts
- range(start, stop, step) and enumerate()
- while with break/continue

```python
for i in range(5):
    print(i)
```

```python
items = ['a','b','c']
for i, val in enumerate(items):
    print(i, val)
```

### While loop

```python
# Initialize
n = 5
factorial = 1
i = 1

while i <= n:
    factorial *= i
    i += 1

print("Factorial of", n, "is", factorial)
```

### break and continue

```python
# countdown and continue/break demo
n = 5
while n > 0:
    if n == 3:
        n -= 1
        continue  # skip printing 3
    if n == 2:
        break     # exit loop when n is 1
    print('n =', n)
    n -= 1
```

## 1. List Overview

- Lists are ordered, zero-indexed, and mutable
- Support slicing and heterogeneous elements

```python
items = [1, 'two', 3.0, ['nested']]
print(type(items))
print(items)
nums = [3, 1, 4, 1, 5]
print('first/last:', nums[0], nums[-1])
print('slice [1:4]:', nums[1:4])
nums.append(9)
```

```python
nums.insert(1, 1.5)
nums.append(1)
nums.remove(1)
popped = nums.pop()
print(nums, popped)
```
