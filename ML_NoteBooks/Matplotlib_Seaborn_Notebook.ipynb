# Matplotlib & Seaborn: The Ultimate Syntax Guide

This notebook is aligned with the tutorial manual.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

## 1. The Mental Model


### Matplotlib: The "Artist" Model
Figure (Canvas) -> Axes (Plot) -> Artists (Lines, Points)


```python
# The Golden Rule: Create Figure and Axes first

x = np.linspace(0, 10, 50)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, label='Sine Wave')
ax.set_title('Matplotlib Mental Model')
ax.legend()
plt.show()
```

### Seaborn: The "Data-Centric" Model
Pass DataFrame, map columns to x, y, hue.


```python
# Create Dummy Data
df = pd.DataFrame({
    'time': x,
    'value': y,
    'category': ['A' if val > 0 else 'B' for val in y]
})

# The Golden Rule: DataFrame + Column Mappings
sns.lineplot(data=df, x='time', y='value', hue='category')
plt.title('Seaborn Mental Model')
plt.show()
```

## 2. Universal Syntax Patterns


### Matplotlib Pattern: plt.plot_type(x, y)


```python
# Explicit X and Y arrays
x_val = [1, 2, 3]
y_val = [4, 5, 6]

plt.figure()
plt.plot(x_val, y_val, color='red', label='Manual Data')
plt.legend()
plt.show()
```

### Seaborn Pattern: sns.plot_type(data=df, x=col, y=col)


```python
# Using DataFrame columns
sns.scatterplot(data=df, x='time', y='value', hue='category')
plt.show()
```

## 3. Essential Methods Cheat Sheet


### Setup & Layout


```python
# Figure size and subplots
fig, ax = plt.subplots(figsize=(10, 4))
ax.text(0.5, 0.5, 'Canvas Area', ha='center')
plt.show()
```

### Labeling & Decorating


```python
x = np.linspace(0, 5, 20)
y = x ** 2

plt.figure()
plt.plot(x, y, label='Growth')
plt.title('My Title')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

## 4. Plot Types Reference


### Relational Plots (Trends & Relationships)


#### Line Plot


```python
# Matplotlib
plt.figure()
plt.plot(df['time'], df['value'])
plt.title('Matplotlib Line')
plt.show()

# Seaborn
sns.lineplot(data=df, x='time', y='value')
plt.title('Seaborn Line')
plt.show()
```

#### Scatter Plot


```python
# Matplotlib
plt.figure()
plt.scatter(df['time'], df['value'])
plt.title('Matplotlib Scatter')
plt.show()

# Seaborn
sns.scatterplot(data=df, x='time', y='value', hue='category')
plt.title('Seaborn Scatter')
plt.show()
```

### Distribution Plots (Spread & Frequency)


#### Histogram


```python
# Matplotlib
plt.figure()
plt.hist(df['value'], bins=10)
plt.title('Matplotlib Histogram')
plt.show()

# Seaborn
sns.histplot(data=df, x='value', kde=True)
plt.title('Seaborn Histogram')
plt.show()
```

#### Box Plot


```python
# Seaborn Box
sns.boxplot(data=df, x='category', y='value')
plt.title('Seaborn Box Plot')
plt.show()
```

### Categorical Plots (Comparison)


#### Bar Chart


```python
# Create categorical summary using groupby (Pandas mental model)
# Avoiding np.unique or complex numpy ops not covered in tutorial
summary_df = df.groupby('category')['value'].mean().reset_index()

# Matplotlib
plt.figure()
plt.bar(summary_df['category'], summary_df['value'])
plt.title('Matplotlib Bar')
plt.show()

# Seaborn
sns.barplot(data=df, x='category', y='value') # Autosummarizes
plt.title('Seaborn Bar')
plt.show()
```

### Matrix Plots (Correlations)


#### Heatmap


```python
# Correlation Matrix
# Selecting only numeric columns
numeric_df = df[['time', 'value']]
corr_matrix = numeric_df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Exploration


#### Pair Plot


```python
sns.pairplot(df, hue='category')
plt.show()
```
